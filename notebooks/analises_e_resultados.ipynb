```python
# ==============================================================================
# SEÇÃO 5.1.3: ANÁLISE DA BASE DE MENÇÕES - PERCEPÇÃO PÚBLICA (VERSÃO FINAL LADO A LADO)
# ==============================================================================
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

print("Iniciando a análise da base de dados de menções (df_mencoes)...")

# --- 5.1.3.1. Análise de Sub-Temas (Feature Engineering no Texto) ---
# (A função classificar_sub_tema permanece a mesma)
def classificar_sub_tema(row):
    tema = row['tema_conteudo']
    conteudo = row['conteudo_post'].lower()
    if tema == 'reclamacao':
        if 'fila' in conteudo or 'espera' in conteudo: return 'Reclamação (Operação/Fila)'
        if 'pessoal' in conteudo or 'atendimento' in conteudo: return 'Reclamação (Atendimento)'
        if 'decepcionado' in conteudo or 'fraca' in conteudo: return 'Reclamação (Qualidade/Experiência)'
        return 'Reclamação (Geral)'
    elif tema == 'elogio':
        if 'brinde' in conteudo: return 'Elogio (Brinde)'
        if 'organização' in conteudo or 'atendimento' in conteudo: return 'Elogio (Atendimento)'
        if 'sensacional' in conteudo or 'produto' in conteudo: return 'Elogio (Qualidade/Experiência)'
        return 'Elogio (Geral)'
    elif tema == 'neutro':
        if 'alguém sabe' in conteudo or '?' in conteudo: return 'Dúvida'
        if 'patrocinadores' in conteudo: return 'Menção Informativa'
        return 'Menção Neutra'
    return 'Outro'

if 'sub_tema' not in df_mencoes.columns:
    df_mencoes['sub_tema'] = df_mencoes.apply(classificar_sub_tema, axis=1)
    print("\nColuna 'sub_tema' criada com sucesso.")


# --- 5.1.3.2. Gráfico 7: Análise do Sentimento e Detalhamento dos Temas ---
# Criando a figura e a grade de layout
fig = plt.figure(figsize=(24, 10))
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1.5]) # 1 linha, 2 colunas, a da direita é 50% mais larga

# --- Gráfico 7.1: Distribuição do Sentimento (Gráfico da Esquerda) ---
ax0 = plt.subplot(gs[0])
sentiment_counts = df_mencoes['sentimento_calculado'].value_counts()
sentiment_labels = sentiment_counts.index
sentiment_colors = ['mediumseagreen' if label == 'Positivo' else 'crimson' if label == 'Negativo' else 'royalblue' for label in sentiment_labels]

ax0.pie(
    sentiment_counts,
    labels=sentiment_labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=sentiment_colors,
    wedgeprops={"edgecolor":"white", 'linewidth': 1, 'antialiased': True},
    textprops={'fontsize': 14}
)
ax0.set_title('Gráfico 7.1: Distribuição do Sentimento Geral (NLP)', fontsize=16)


# --- Gráfico 7.2: Distribuição dos Sub-Temas (Gráfico da Direita) ---
ax1 = plt.subplot(gs[1])
sns.countplot(
    data=df_mencoes,
    y='sub_tema',
    order=df_mencoes['sub_tema'].value_counts().index,
    palette='viridis',
    ax=ax1
)

# Adicionando os rótulos de contagem ao lado de cada barra
for p in ax1.patches:
    width = p.get_width()
    ax1.text(width + 3, # Posição x
             p.get_y() + p.get_height() / 2, # Posição y
             f'{int(width)}', # O texto
             va='center',
             fontsize=12)

ax1.set_title('Gráfico 7.2: Detalhamento dos Temas das Menções', fontsize=16)
ax1.set_xlabel('Contagem de Menções', fontsize=14)
ax1.set_ylabel('Sub-Tema', fontsize=14)
ax1.set_xlim(0, df_mencoes['sub_tema'].value_counts().max() * 1.15)


# Adicionando um título geral e ajustando o layout
plt.suptitle('Gráfico 7: Análise da Percepção Pública sobre a Marca', fontsize=20, y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# ==============================================================================
# SEÇÃO 5.2: PRIMEIRAS CONEXÕES - ANÁLISE EXPLORATÓRIA E CAUSAL (NUMERAÇÃO CORRIGIDA)
# ==============================================================================

print("Iniciando a análise exploratória para identificar relações entre as variáveis...")

# Filtrando o dataframe principal para focar apenas nas interações com a marca em estudo
segmento_alvo = "Bebida Alcoólica"
df_marca_foco = df_principais[df_principais['segmento_interagido'] == segmento_alvo].copy()

print(f"\nAnálise focada no segmento: '{segmento_alvo}'. Total de interações: {len(df_marca_foco)}")


# --- 5.2.1. Gráfico 8: Investigando os Drivers da Satisfação com a Marca ---
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

# Gráfico de Dispersão para Tempo de Espera vs. Satisfação
sns.regplot(
    data=df_marca_foco, x='tempo_de_espera', y='satisfacao_segmento', ax=axes[0],
    scatter_kws={'alpha':0.3, 'color': 'orangered'}, line_kws={'color': 'darkred', 'linewidth': 3}
)
axes[0].set_title('Gráfico 8.1: Impacto do Tempo de Espera na Satisfação com a Marca')
axes[0].set_xlabel('Tempo de Espera (minutos)')
axes[0].set_ylabel('Nota de Satisfação com o Segmento')

# Boxplot para Qualidade da Experiência vs. Satisfação
sns.boxplot(
    data=df_marca_foco, x='avaliacao_experiencia_ativacao', y='satisfacao_segmento', ax=axes[1], palette='YlGn'
)
axes[1].set_title('Gráfico 8.2: Impacto da Qualidade da Experiência na Satisfação')
axes[1].set_xlabel('Avaliação da Experiência na Ativação (1 a 5)')
axes[1].set_ylabel('Nota de Satisfação com o Segmento')

plt.suptitle('Gráfico 8: Análise dos Fatores de Satisfação com a Marca', fontsize=16, y=1.03)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


# --- 5.2.2. Gráfico 9: O Impacto do Brinde ---
df_marca_foco['recebeu_brinde'] = df_marca_foco['avaliacao_brinde'] > 0

plt.figure(figsize=(10, 7))
sns.boxplot(
    data=df_marca_foco, x='recebeu_brinde', y='satisfacao_segmento', palette=['lightcoral', 'mediumseagreen']
)
plt.title('Gráfico 9: Comparação da Satisfação com e sem Recebimento de Brinde')
plt.xlabel('Recebeu Brinde?')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.ylabel('Nota de Satisfação com o Segmento')
plt.show()


# --- 5.2.3. Gráfico 10: Satisfação e Lealdade (Recomendação) ---
df_marca_foco['faixa_satisfacao'] = pd.cut(df_marca_foco['satisfacao_segmento'], bins=[0, 1, 2, 3, 4, 5], right=True)
prob_recomendar = df_marca_foco.groupby('faixa_satisfacao', observed=True)['recomendaria_segmento'].value_counts(normalize=True).unstack()
prob_recomendar_sim = prob_recomendar.get(True, pd.Series(0, index=prob_recomendar.index)) * 100 # .get() para evitar erro se ninguém recomendar

plt.figure(figsize=(12, 7))
ax = sns.barplot(x=prob_recomendar_sim.index.astype(str), y=prob_recomendar_sim.values, palette='coolwarm')
ax.bar_label(ax.containers[0], fmt='%.1f%%')

plt.title('Gráfico 10: Probabilidade de Recomendar a Marca por Nível de Satisfação')
plt.xlabel('Faixa de Satisfação com o Segmento')
plt.ylabel('Probabilidade de Recomendar (%)')
plt.ylim(0, 110)
plt.show()
```

```python
# ==============================================================================
# SEÇÃO 5.2.2a: ANÁLISE INFERENCIAL - TESTE DE HIPÓTESE PARA O IMPACTO DO BRINDE
# ==============================================================================
from scipy.stats import ttest_ind

print("\nExecutando Teste de Hipótese (Teste T) para o impacto do brinde...")

# Separando os grupos
satisfacao_com_brinde = df_marca_foco[df_marca_foco['recebeu_brinde'] == True]['satisfacao_segmento']
satisfacao_sem_brinde = df_marca_foco[df_marca_foco['recebeu_brinde'] == False]['satisfacao_segmento']

# Executando o Teste T
t_stat, p_value = ttest_ind(satisfacao_com_brinde, satisfacao_sem_brinde, nan_policy='omit')

print("\n--- Resultados do Teste T de Amostras Independentes ---")
print(f"Estatística T: {t_stat:.4f}")
print(f"P-valor (p-value): {p_value:.4f}")

# Interpretando o resultado
alpha = 0.05 # Nível de significância padrão
if p_value < alpha:
    print(f"\nConclusão: Como o p-valor ({p_value:.4f}) é menor que {alpha}, rejeitamos a hipótese nula.")
    print("A diferença na satisfação média entre os grupos COM e SEM brinde é ESTATISTICAMENTE SIGNIFICATIVA.")
else:
    print(f"\nConclusão: Como o p-valor ({p_value:.4f}) é maior que {alpha}, não podemos rejeitar a hipótese nula.")
    print("A diferença observada na satisfação média entre os grupos não é estatisticamente significativa.")
```

```python
# ==============================================================================
# SEÇÃO 5.2.4: ANÁLISE DE NÃO-RELAÇÃO - GASTO VS. SATISFAÇÃO
# ==============================================================================
# Objetivo: Investigar se existe uma relação causal entre o valor gasto
# pelo participante e sua satisfação com a marca.

print("\nInvestigando a relação entre Valor Gasto e Satisfação com a Marca...")

# --- Gráfico 11: Análise de Correlação entre Gasto e Satisfação ---

plt.figure(figsize=(12, 7))

# Usamos um scatter plot para visualizar a relação
sns.scatterplot(
    data=df_marca_foco,
    x='valor_gasto',
    y='satisfacao_segmento',
    alpha=0.3 # Deixa os pontos transparentes para ver a densidade
)

# Calculando a correlação de Pearson para quantificar a relação
correlacao = df_marca_foco['valor_gasto'].corr(df_marca_foco['satisfacao_segmento'])

plt.title(f'Gráfico 11: Relação entre Valor Gasto e Satisfação com a Marca\nCorrelação de Pearson: {correlacao:.3f}')
plt.xlabel('Valor Gasto no Dia (R$)')
plt.ylabel('Nota de Satisfação com o Segmento')
plt.show()

print("\n--- Conclusão da Análise de Correlação ---")
print(f"O coeficiente de correlação de Pearson entre 'valor_gasto' e 'satisfacao_segmento' é {correlacao:.3f}.")
if abs(correlacao) < 0.2:
    print("Este é um valor muito próximo de zero, indicando uma correlação linear muito fraca ou inexistente.")
else:
    print("Há uma correlação moderada ou forte que precisa ser investigada.")
```

```python
# ==============================================================================
# SEÇÃO 5.3: O MODELO INTEGRADO - ANÁLISE PREDITIVA E PRESCRITIVA
# ==============================================================================

# --- 5.3.1. Análise Preditiva: Quantificando os Drivers da Recomendação ---
# Objetivo: Usar Machine Learning para descobrir quais fatores mais influenciam
# a probabilidade de um cliente recomendar a marca.

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import plotly.express as px # Usaremos Plotly para um gráfico interativo

print("Iniciando a análise preditiva com Random Forest...")

# --- 1. Preparação dos Dados (Feature Engineering) ---

# Usaremos o mesmo dataframe focado na marca que criamos na seção 5.2
# df_marca_foco já está filtrado para 'Bebida Alcoólica'

# Selecionando as colunas que serão usadas como 'features' (preditores)
features = [
    'tempo_de_espera',
    'avaliacao_experiencia_ativacao',
    'avaliacao_brinde',
    'satisfacao_segmento',
    'idade',
    'valor_gasto',
    'frequencia_festival',
    'dia_festival',
    'genero'
]

# Definindo a variável 'target' (o que queremos prever)
target = 'recomendaria_segmento'

# Criando o dataframe final para o modelo
df_modelo = df_marca_foco[features + [target]].copy()

# Convertendo variáveis categóricas em numéricas usando One-Hot Encoding
# Isso cria novas colunas para cada categoria (ex: 'genero_Masculino', 'genero_Feminino')
df_modelo_encoded = pd.get_dummies(df_modelo, columns=['frequencia_festival', 'dia_festival', 'genero'], drop_first=True)

print("\nDados preparados para o modelo.")
display(df_modelo_encoded.head())


# --- 2. Treinamento do Modelo de Classificação ---

# Separando os dados em features (X) e target (y)
X = df_modelo_encoded.drop(target, axis=1)
y = df_modelo_encoded[target]

# Dividindo os dados em conjuntos de treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Criando e treinando o classificador Random Forest
# n_estimators é o número de árvores na floresta. random_state para reprodutibilidade.
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, oob_score=True)
rf_classifier.fit(X_train, y_train)

print("\nModelo Random Forest treinado com sucesso.")


# --- 3. Avaliação da Performance do Modelo ---

# Fazendo previsões no conjunto de teste
y_pred = rf_classifier.predict(X_test)

# Calculando e imprimindo a acurácia
accuracy = accuracy_score(y_test, y_pred)
oob_accuracy = rf_classifier.oob_score_

print("\n--- Performance do Modelo de Classificação ---")
print("="*45)
print(f"Acurácia no conjunto de teste: {accuracy * 100:.2f}%")
print(f"Acurácia Out-of-Bag (OOB): {oob_accuracy * 100:.2f}%")
print("="*45)
# OOB score é uma forma de validação cruzada interna do Random Forest, muito confiável.


# --- 4. Análise de Feature Importance (O Insight Principal) ---
# Extraindo a importância de cada feature do modelo treinado
feature_importances = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rf_classifier.feature_importances_
}).sort_values('importance', ascending=False)

print("\nImportância de cada variável para prever a recomendação:")
display(feature_importances)

# Gráfico 12: Visualização da Importância das Features
fig = px.bar(
    feature_importances,
    x='importance',
    y='feature',
    orientation='h',
    title='Gráfico 12: Importância das Variáveis na Previsão de Recomendação da Marca',
    labels={'importance': 'Importância Relativa', 'feature': 'Variável'},
    text='importance'
)
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.update_traces(texttemplate='%{text:.3f}', textposition='outside')
fig.show()
```

```python
# ==============================================================================
# SEÇÃO 5.3.7: SIMULAÇÃO PRESCRITIVA ALTERNATIVA (IMPACTO DO BRINDE)
# ==============================================================================
# Objetivo: Quantificar o impacto de uma melhoria na qualidade dos brindes
# na probabilidade de recomendação da marca.

print("\nIniciando a simulação 'what-if' para a avaliação do brinde...")

# --- 1. Identificar o Segmento de Foco ---
# Vamos focar nos clientes que receberam brinde, mas não o avaliaram com nota máxima.
# Ou seja, aqueles com avaliacao_brinde entre 1 e 4.
segmento_brinde_melhoria_idx = X_test[ (X_test['avaliacao_brinde'] > 0) & (X_test['avaliacao_brinde'] < 5) ].index

# --- 2. Estabelecer a Baseline PARA ESTE GRUPO ---
if not segmento_brinde_melhoria_idx.empty:
    probabilidades_base_brinde = rf_classifier.predict_proba(X_test.loc[segmento_brinde_melhoria_idx])[:, 1]
    media_prob_base_brinde = probabilidades_base_brinde.mean()
    print(f"Cenário Base (Clientes com Brinde 'OK'): A prob. média de recomendar é de {media_prob_base_brinde * 100:.2f}%")

    # --- 3. Criar o Cenário Simulado ---
    # Hipótese: "E se todos os brindes fossem excelentes (nota 5)?"
    X_test_simulado_brinde = X_test.copy()
    X_test_simulado_brinde.loc[segmento_brinde_melhoria_idx, 'avaliacao_brinde'] = 5
    print("\nCenário Simulado: Avaliação de todos os brindes melhorada para a nota máxima (5).")

    # --- 4. Fazer Novas Previsões e Avaliar o Impacto ---
    probabilidades_simulado_brinde = rf_classifier.predict_proba(X_test_simulado_brinde.loc[segmento_brinde_melhoria_idx])[:, 1]
    media_prob_simulado_brinde = probabilidades_simulado_brinde.mean()

    aumento_absoluto = (media_prob_simulado_brinde - media_prob_base_brinde) * 100
    aumento_relativo = ((media_prob_simulado_brinde - media_prob_base_brinde) / media_prob_base_brinde) * 100

    print(f"Cenário Simulado (Clientes com Brinde 'OK'): A nova prob. média seria de {media_prob_simulado_brinde * 100:.2f}%")
    print("\n--- IMPACTO ESTRATÉGICO DA MELHORIA NOS BRINDES ---")
    print("="*55)
    print(f"Aumento Absoluto na Prob. de Recomendação: +{aumento_absoluto:.2f} p.p.")
    print(f"Aumento Relativo na Prob. de Recomendação: +{aumento_relativo:.2f}%")
    print("="*55)
else:
    print("\nNão foram encontrados participantes no conjunto de teste que se encaixam nos critérios da simulação do brinde.")
```

```python
# ==============================================================================
# SEÇÃO 5.3.4: ANÁLISE PRESCRITIVA (SIMULAÇÃO "WHAT-IF")
# ==============================================================================
# Objetivo: Utilizar o modelo Random Forest treinado como um simulador para
# quantificar o impacto de uma mudança operacional na recomendação da marca.

print("\nIniciando a análise de simulação 'what-if'...")

# --- 1. Estabelecer o Cenário Base (Baseline) ---
# Primeiro, calculamos a probabilidade média de recomendação no conjunto de teste original.
# O método predict_proba retorna as probabilidades para cada classe [Não, Sim].
# Queremos a probabilidade da classe 'Sim', que é a segunda coluna [:, 1].
probabilidades_base = rf_classifier.predict_proba(X_test)[:, 1]
media_prob_base = probabilidades_base.mean()

print(f"Cenário Base: A probabilidade média de um cliente recomendar a marca é de {media_prob_base * 100:.2f}%")


# --- 2. Criar o Cenário Hipotético ---
# Hipótese: "E se conseguíssemos reduzir o tempo de espera em 50%?"

# Criamos uma cópia do conjunto de teste para não alterar o original
X_test_simulado = X_test.copy()

# Aplicando a mudança na variável 'tempo_de_espera'
reducao_percentual = 0.50 # 50%
X_test_simulado['tempo_de_espera'] = X_test_simulado['tempo_de_espera'] * (1 - reducao_percentual)

print(f"\nCenário Simulado: Redução de {reducao_percentual * 100:.0f}% no tempo de espera aplicada.")


# --- 3. Fazer Novas Previsões e Avaliar o Impacto ---
# Usamos o mesmo modelo treinado para prever o resultado no cenário simulado
probabilidades_simulado = rf_classifier.predict_proba(X_test_simulado)[:, 1]
media_prob_simulado = probabilidades_simulado.mean()

# Calculando o impacto da mudança
aumento_absoluto = (media_prob_simulado - media_prob_base) * 100
aumento_relativo = ((media_prob_simulado - media_prob_base) / media_prob_base) * 100

print(f"Cenário Simulado: A nova probabilidade média de recomendação seria de {media_prob_simulado * 100:.2f}%")
print("\n--- IMPACTO ESTRATÉGICO QUANTIFICADO ---")
print("="*45)
print(f"Aumento Absoluto na Prob. de Recomendação: +{aumento_absoluto:.2f} pontos percentuais")
print(f"Aumento Relativo na Prob. de Recomendação: +{aumento_relativo:.2f}%")
print("="*45)
```
